```python
from matplotlib import pyplot as pt
```

```python
pt.pie(employee['department_id'].value_counts(), labels =employee['department_id'].unique(), autopct="%0.1f%%")
pt.show()
```

```python
#FOR BAR GRAPH
ax = employee.department_id.value_counts().plot(kind="bar")
ax.set_xlabel("Department ID")
ax.set_ylabel("Number of Employees")
```

```python
#FOR SUBPLOTS
pt.subplot(1,2,1)
ax = employee.department_id.value_counts().plot(kind="bar")
ax.set_title("Number of Employees/Department")
ax.set_xlabel("Department ID")
ax.set_ylabel("Number of Employees")

pt.subplot(1,2,2)
employee.department_id.value_counts().plot(kind="pie", autopct="%0.1f%%")
```
